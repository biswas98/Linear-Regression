```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

dataSet = pd.read_csv("Data Sets/Salary_Data-1.csv")
dataSet
```

```python
dataSet.plot(kind='scatter',x='YearsExperience',y='Salary')
plt.show()
```

```python
x = dataSet.drop('Salary', axis='columns')
y = dataSet.drop('YearsExperience', axis='columns')
```

```python
x
```

```python
y
```

```python
LM = linear_model.LinearRegression()
model = LM.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x,y) 
```

```python
#predict new salary value
x_new=7
y_Predicted = model.predict([[x_new]])
y_Predicted
```

```python
x_new=([1,3,5])
x_new=pd.DataFrame(x_new)
y_new=model.predict(x_new)
y_new=pd.DataFrame(y_new)
df=pd.concat([x_new,y_new], axis=1, keys=['New Experience','New Salary'])
df
```
